### 2. Start Time

```python
df.Start_Time
```

```python
## Converting to the desired datetime format

df.Start_Time= pd.to_datetime(df.Start_Time)
```

```python
df.Start_Time
```

```python
## Starting hours of the accidents

Start_hour= df.Start_Time.dt.hour
Start_hour
```

```python
## Ploting a percentage ditribution chart of the Starting hours of accidents

plt.figure(figsize=(12,6))
sns.distplot(Start_hour, bins= 24, norm_hist=True);
```

```python
## Starting day of the week of accidents
Start_dayofweek= df.Start_Time.dt.dayofweek
Start_dayofweek
```

```python
## Ploting a percentage ditribution chart of the day of the week of accidents

plt.figure(figsize=(12,6))
sns.distplot(Start_dayofweek, bins= 7, kde=False, norm_hist=True);
```

```python
## Sunday and Monday's start time of accidents

Sunday_start_time= df.Start_Time[df.Start_Time.dt.dayofweek== 6]
Monday_start_time= df.Start_Time[df.Start_Time.dt.dayofweek== 0]
```

```python
## Ploting a percentage ditribution chart of Sundays hour wise accidents


plt.figure(figsize=(12,6))
sns.distplot(Sunday_start_time.dt.hour, bins= 24, kde=False, norm_hist=True);
```

```python
## Ploting a percentage ditribution chart of Monday's hour wise accidents


plt.figure(figsize=(12,6))
sns.distplot(Monday_start_time.dt.hour, bins= 24, kde= False, norm_hist= True);
```

```python
## Ploting a percentage ditribution chart of monthwise wise accidents

plt.figure(figsize=(12, 6))
sns.distplot(df.Start_Time.dt.month, bins= 12, kde=False, norm_hist=True);
```

```python
## Fetching Start_Time for the year 2019
df_2019= df[df.Start_Time.dt.year== 2019]
df_2019
```

```python
## Ploting a percentage of monthwise ditribution for Start_Time, 2019 of accidents

plt.figure(figsize=(12,6))
df_2019= df[df.Start_Time.dt.year== 2019]
sns.distplot(df_2019.Start_Time.dt.month, bins= 12, kde= False, norm_hist=True);
```

```python
## Ploting a percentage of monthwise ditribution for Start_Time, 2016 of accidents

plt.figure(figsize=(12,6))
df_2016= df[df.Start_Time.dt.year== 2016]
sns.distplot(df_2016.Start_Time.dt.month, bins= 12, kde= False, norm_hist=True);
```

```python
## Ploting a percentage of monthwise ditribution for Start_Time, 2020 of accidents

plt.figure(figsize=(12,6))
df_2020= df[df.Start_Time.dt.year== 2020]
sns.distplot(df_2020.Start_Time.dt.month, bins= 12, kde= False, norm_hist=True);
```
